````
#!fsharp
let parse (input:string) = 
    input.Split("\n") 
    |> Array.map (fun x -> x.Trim()) 
    |> Array.filter (fun x -> x <> "") 
let testMap = parse testInput
````

````
#!fsharp
let getMapTile ( x, y ) ( map: string[]) =
    if x < 1 || y < 1 || y > map.Length then 
        None
    else 
        let width = map.[0].Length
        Some map.[y-1].[ (x-1) % width ]

let walk (x,y) (sx, sy) fn init map = 
    let rec inner x y acc =
        let ch = getMapTile (x, y) map
        match ch with
        | Some c -> 
            inner (x+sx) (y+sy) (fn c acc)
        | None -> 
            acc
    inner x y init
                
let countTrees (sx, sy) map = 
    walk (1,1) (sx,sy) (fun ch acc ->
        if ch = '#' then acc+1 else acc
    ) 0 map

countTrees (3, 1) testMap
````

Time to check the rest of the slopes - you need to minimize the probability of a sudden arboreal stop, after all.

Determine the number of trees you would encounter if, for each of the following slopes, you start at the top-left corner and traverse the map all the way to the bottom:

```
    Right 1, down 1.
    Right 3, down 1. (This is the slope you already checked.)
    Right 5, down 1.
    Right 7, down 1.
    Right 1, down 2.
```

In the above example, these slopes would find 2, 7, 3, 4, and 2 tree(s) respectively; multiplied together, these produce the answer 336.

What do you get if you multiply together the number of trees encountered on each of the listed slopes?

````
#!fsharp
let tryMultiple map =
    let s1 = double ( countTrees (1, 1) map )
    let s2 = double ( countTrees (3, 1) map )
    let s3 = double ( countTrees (5, 1) map )
    let s4 = double ( countTrees (7, 1) map )
    let s5 = double ( countTrees (1, 2) map )
    s1 * s2 * s3 * s4 * s5

tryMultiple testMap
````
